```r
library(tidyverse)
```

```r
data1 <- read_csv("res.csv")
data2 <- read_csv("res2.csv")
data_lorem <- read_csv("res_lorem.csv")
```

```r
col_names <- c("input_size","entropy","output_size","time")
```

```r
colnames(data1)<-col_names
data1 <- data1%>%
mutate(percent_compressed=output_size/input_size*100)
head(data1)
```

```r
colnames(data2)<-col_names
data2 <- data2%>%
mutate(percent_compressed=output_size/input_size*100)
head(data2)
```

```r
colnames(data_lorem)<-col_names
data_lorem <- data_lorem%>%
mutate(percent_compressed=output_size/input_size*100)
head(data_lorem)
```

```r
ggplot(data1,aes(x=input_size,y=percent_compressed))+
geom_smooth(se=FALSE)+
geom_hline(aes(yintercept=mean(entropy*12.5)),linetype = "longdash",color="red")+
geom_text(aes(x=0,y=mean(entropy*12.5)+3,label="75%"))+
labs(title="Percent compressed vs Input Size (purely random text)")+
ylab("Percent Compressed %")+
xlab("Input Size (bytes)")
```

```r
ggplot(data2,aes(x=input_size,y=percent_compressed))+
geom_smooth(se=FALSE)+
geom_hline(aes(yintercept=mean(entropy*12.5)),linetype = "longdash",color="red")+
geom_text(aes(x=0,y=mean(entropy*12.5)+3,label="75%"))+
labs(title="Percent compressed vs Input Size (purely random text)")+
ylab("Percent Compressed %")+
xlab("Input Size (bytes)")
```

```r
ggplot(data_lorem,aes(x=input_size,y=percent_compressed))+
geom_smooth(se=FALSE)+
geom_hline(aes(yintercept=mean(entropy*12.5)),linetype = "longdash",color="red")+
geom_text(aes(x=0,y=mean(entropy*12.5)+3,label="50%"))+
labs(title="Percent compressed vs Input Size (benchmark english texts)")+
ylab("Percent Compressed %")+
xlab("Input Size (bytes)")
```

```r
ggplot(data1,aes(x=input_size,y=time))+
geom_point(alpha=0.1)+
geom_smooth(method="lm",se=FALSE,color="red")+
labs(title="Time taken to compress vs Input Size (purely random text)")+
ylab("Time taken (sec)")+
xlab("Input Size (bytes)")
```

```r
ggplot(data2,aes(x=input_size,y=time))+
geom_point(alpha=0.1)+
geom_smooth(method="lm",se=FALSE,color="red")+
labs(title="Time taken to compress vs Input Size (purely random text)")+
ylab("Time taken (sec)")+
xlab("Input Size (bytes)")
```

```r
ggplot(data_lorem,aes(x=input_size,y=time))+
geom_point(alpha=0.1)+
geom_smooth(method="lm",se=FALSE,color="red")+
labs(title="Time taken to compress vs Input Size (benchmark english texts)")+
ylab("Time taken (sec)")+
xlab("Input Size (bytes)")
```

```r
ggplot(data1,aes(x=input_size,y=output_size))+
geom_point(alpha=0.1)+
geom_smooth(method="lm",se=FALSE,color="red")+
labs(title="Output Size vs Input Size (purely random text)")+
ylab("Output Size (bytes)")+
xlab("Input Size (bytes)")
#Too much overplotting :((
```

```r
ggplot(data2,aes(x=input_size,y=output_size))+
geom_point(alpha=0.1)+
geom_smooth(method="lm",se=FALSE,color="red")+
labs(title="Output Size vs Input Size (purely random text)")+
ylab("Output Size (bytes)")+
xlab("Input Size (bytes)")
```

```r
ggplot(data_lorem,aes(x=input_size,y=output_size))+geom_point(alpha=0.1)+
geom_smooth(method="lm",se=FALSE,color="red")+
labs(title="Output Size vs Input Size benchmark english texts)")+
ylab("Output Size (bytes)")+
xlab("Input Size (bytes)")
```

```r
combine <- rbind(data1,data_lorem)
head(combine)
```

```r
combine%>%
filter(percent_compressed<=150)%>%
ggplot(aes(x=entropy,y=percent_compressed,color=input_size))+
geom_point(alpha=0.1)+
scale_color_gradient(name="Input size",low="blue", high="red")+
labs(title="Percent compressed vs Entropy")+
ylab("Percent Compressed %")+
xlab("Entropy")
```
